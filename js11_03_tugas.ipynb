Code :


```python
import numpy as np
import pandas as pd

df = pd.read_csv('data/wbc.csv')

# slice kolom terakhir --> Unnamed column
df = df.iloc[:,:-1]

df.head()
```

```python
X = df.iloc[:, 2:]
y = df['diagnosis']

X.head()
```

```python
from sklearn.model_selection import RepeatedKFold

cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)
```

```python
import time
start = time.time()

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

dt = DecisionTreeClassifier()

params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11))
}

rs = RandomizedSearchCV(dt, param_distributions=params,
                        n_iter=10, scoring="accuracy", random_state=0,
                        n_jobs=-1, cv=10, return_train_score=True)

rs.fit(X, y)

print("Optimal hyperparameter combination:", rs.best_params_)
print()
print("Hasil accuracy score:",
      rs.best_score_)
rs.best_estimator_.fit(X, y)
y_pred = rs.best_estimator_.predict(X) # Predictions
y_true = y # True values

end = time.time()
diff = end - start
print('Execution time of Random Search (in Seconds):', diff)
print()
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
  
diagnosis = load_breast_cancer()
  
print(diagnosis.keys())
```

```python
df = pd.read_csv('data/wbc.csv')
df.head()
```

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
df_feat = pd.DataFrame(diagnosis['data'],
                       columns = diagnosis['feature_names'])
  
df_target = pd.DataFrame(diagnosis['target'], 
                     columns =['Cancer'])
  
print("Feature Variables: ")
print(df_feat.info())
```

```python
print("Dataframe looks like : ")
print(df_feat.head())
```

```python
from sklearn.model_selection import train_test_split
  
X_train, X_test, y_train, y_test = train_test_split(
                        df_feat, np.ravel(df_target),
                test_size = 0.30, random_state = 101)
```

```python
model = SVC()
model.fit(X_train, y_train)
  
predictions = model.predict(X_test)
print(classification_report(y_test, predictions))
```

```python
from sklearn.model_selection import GridSearchCV
  
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
  
print(grid.best_estimator_)
```

```python
grid_predictions = grid.predict(X_test)
  
print(classification_report(y_test, grid_predictions))
```
